```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20, random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
print("intercept_:",lr.intercept_)
print("coef_:", lr.coef_)
```

```python
from sklearn.linear_model import Ridge
R = Ridge(alpha=10)
R.fit(X,y)
print("intercept_:",R.intercept_)
print("coef_:", R.coef_)
```

```python
from sklearn.linear_model import Ridge
R1 = Ridge(alpha=100)
R1.fit(X,y)
print("intercept_:",R1.intercept_)
print("coef_:", R1.coef_)
```

```python
# plt.plot(X,y,'b')
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,R.predict(X),color='green',label='alpha=10')
plt.plot(X,R1.predict(X),color='orange',label='alpha=100')
plt.legend()
```

```python
class MeraReg:
    def __init__(self,alpha=0.1):
        self.alpha = alpha
        self.m = None
        self.b = None

    def fit(self,X_train,y_train):
        pass
        num = 0
        dem = 0
        for i in range(X_train.shape[0]):
            num = num + (y_train[i] - y_train.mean()) * (X_train[i] - X_train.mean()) 
            dem = dem + (X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()) 
        self.m = num/(dem + self.alpha)
        self.b = y_train.mean() - (self.m*X_train.mean())
        print("m",self.m)
        print("b",self.b)
    def predict(self,X_test):
        return (self.m * X_test) + self.b
```

```python
reg = MeraReg(alpha=10)
reg.fit(X,y)
```

```python
# plt.plot(X,y,'b')
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,R.predict(X),color='green',label='alpha=10')
plt.plot(X,R1.predict(X),color='orange',label='alpha=100')
plt.plot(X,reg.predict(X),color='blue',label='alpha=100')
plt.legend()
```
