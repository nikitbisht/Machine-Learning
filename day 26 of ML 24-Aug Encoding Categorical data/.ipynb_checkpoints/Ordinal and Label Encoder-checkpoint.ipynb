```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.3)

X_train.shape, X_test.shape
```

# Orinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Good','Average'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
X_train = pd.DataFrame(X_train,columns=['Grade','Education'])
X_test = pd.DataFrame(X_test,columns=['Grade','Education'])
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(y_train)
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
y_train = pd.DataFrame(y_train,columns=['Purchased'])
y_test = pd.DataFrame(y_test,columns=['Purchased'])
```
