```python
upper_limit = df['Height'].quantile(0.995);
lower_limit = df['Height'].quantile(0.005);
upper_limit,lower_limit
```

```python
df[(df['Height'] >= upper_limit) | (df['Height'] <= lower_limit)]
```

# Trimming

```python
new_df = df[(df['Height'] < upper_limit) & (df['Height'] > lower_limit)]
```

```python
new_df
```

```python
new_df["Height"].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'], orient='h')
```

```python
# Capping --> WInsorization
```

```python
new_ndf = df.copy()
new_ndf['Height'] = np.where(
    df['Height'] >= upper_limit, 
    upper_limit,
    np.where(
        df['Height'] <= lower_limit,
        lower_limit,
        df['Height']
    )
)
```

```python
new_ndf['Height'].describe()
```

```python
sns.distplot(new_ndf['Height'])
```

```python
sns.boxplot(new_ndf['Height'],orient='h')
```
