```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# lets apply Gradient Desent and assuming slope is constant m = 78.35
# and assume stating value of intercept b = 0
y_pred = ((78.35 * X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label="OLS")
plt.plot(X,y_pred,color = 'green',label='b-0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# let take learning rate = 0.1
lr = 0.1
step_size = lr*loss_slope
step_size
```

```python
#Calculate new intecept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label="OLS")
plt.plot(X,y_pred1,color = 'green',label='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label='b-0')
plt.legend()
plt.show()
```

```python
# 2nd Iteration 
loss_slope = -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = (m*X + b).reshape(4)
y_pred2
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label="OLS")
plt.plot(X,y_pred2,color = 'black',label='b = {}'.format(b))
plt.plot(X,y_pred1,color = 'green',label='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label='b-0')
plt.legend()
plt.show()
```

```python
# Iteration 3
```

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = (m*X + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label="OLS")
plt.plot(X,y_pred3,color = 'purple',label='b = {}'.format(b))
plt.plot(X,y_pred2,color = 'black',label='b = {}'.format(b))
plt.plot(X,y_pred1,color = 'green',label='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label='b-0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1

epochs = 100
for i in range(epochs):
    loss_slope = -2 * np.sum( y - X.ravel() - b)
    b = b - (loss_slope*lr)
    y_pred = m*X + b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
