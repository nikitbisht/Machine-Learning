```python
import pandas as pd
import numpy as np

import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
X=df.iloc[:,1:3]
Y = df.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
plt.figure(figsize=(6,4))
sns.distplot(X_train['Age'])
plt.title('Age PDF')
plt.show()
plt.figure(figsize=(6,4))
stats.probplot(X_train['Age'],dist='norm',plot=plt)
plt.title('Age QQ plot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title('Fare Pda')
sns.distplot(X_train['Fare'])
plt.show()

plt.figure(figsize=(6,4))
stats.probplot(X_train['Fare'],dist='norm',plot=plt)
plt.title('Fare QQ plot')
plt.show()
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)

y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)

print('Accuracy LR',accuracy_score(y_pred1, y_test))
print('Accuracy DT',accuracy_score(y_pred2, y_test))
```

```python
trf = FunctionTransformer(func=np.log1p)
```

```python
X_train_transformed = trf.fit_transform(X_train)
X_test_transformed = trf.fit_transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train_transformed,y_train)
clf2.fit(X_train_transformed,y_train)

y_pred1 = clf1.predict(X_test_transformed)
y_pred2 = clf2.predict(X_test_transformed)

print('Accuracy LR',accuracy_score(y_pred1, y_test))
print('Accuracy DT',accuracy_score(y_pred2, y_test))
```

```python
X_transformed = trf.fit_transform(X)

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

print("LR", np.mean(cross_val_score(clf1,X_transformed,Y,scoring='accuracy',cv=10)))
print("DT", np.mean(cross_val_score(clf2,X_transformed,Y,scoring='accuracy',cv=10)))
```

```python
plt.figure(figsize=(6,4))
stats.probplot(X_train['Fare'],dist='norm',plot=plt)
plt.title('Before Log Fare QQ plot')
plt.show()

stats.probplot(X_train_transformed['Fare'],dist='norm',plot=plt)
plt.title('After Log Fare QQ plot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
stats.probplot(X_train['Age'],dist='norm',plot=plt)
plt.title('Before Log Age QQ plot')
plt.show()

stats.probplot(X_train_transformed['Age'],dist='norm',plot=plt)
plt.title('After Log Age QQ plot')
plt.show()
```

```python
trf1 = ColumnTransformer(transformers=[('log',FunctionTransformer(func=np.log1p),['Fare'])],remainder='passthrough')

X_train_transformed2 = trf1.fit_transform(X_train)
X_test_transformed2 = trf1.fit_transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train_transformed2,y_train)
clf2.fit(X_train_transformed2,y_train)

y_pred1 = clf1.predict(X_test_transformed2)
y_pred2 = clf2.predict(X_test_transformed2)

print('Accuracy LR',accuracy_score(y_pred1, y_test))
print('Accuracy DT',accuracy_score(y_pred2, y_test))
```
