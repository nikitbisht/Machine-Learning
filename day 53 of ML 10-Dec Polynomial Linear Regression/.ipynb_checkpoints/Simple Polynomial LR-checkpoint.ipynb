```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
X = 6* np.random.rand(200,1)-3
y = 0.8 * X**2 + 0.9 * X+2+np.random.randn(200,1)
```

```python
plt.plot(X,y,'b.')
plt.xlabel("X")
plt.ylabel("y")
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X,y,'b.')
plt.xlabel("X")
plt.ylabel("y")
```

```python
# include_bias = flase means x^0 removed
poly = PolynomialFeatures(degree=2,include_bias=True)

X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.fit_transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
lr = LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred = lr.predict(X_test_trans)
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_new,y_new , "r-",linewidth=2,label="predictions")
plt.plot(X_train,y_train,"b.",label="Training points")
plt.plot(X_test,y_test,"g.",label="testing Points")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
```

```python
def polynomial_regressioin(degree):
    X_new = np.linspace(-3,3,100).reshape(100,1)
    X_new_poly = poly.transform(X_new)
    
    poly_freture= PolynomialFeatures(degree=degree,include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ('poly_regerssor',poly_freture),
        ('std_scaler',std_scaler),
        ('lin_reg',lin_reg)
    ])
    
    poly_regression.fit(X,y)
    y_newbig = poly_regression.predict(X_new)
    plt.plot(X_new,y_newbig,'r',label='Degree' + str(degree) , linewidth = 2)
    plt.plot(X_train,y_train,'b.',linewidth=3)
    plt.plot(X_test,y_test,'g.',linewidth=3)
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-3,3,0,10])
    plt.show()
    
```

```python
polynomial_regressioin(degree=300)
```
