```python
import numpy as np
import pickle
import pandas as pd

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectKBest, chi2
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# lets Plan

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']),
                                                    df['Survived'],
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
X_train.head()
```

```python
# Imputation Transformer

trf1 = ColumnTransformer(transformers=[
    ('imput_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
X_traformed = trf1.fit_transform(X_train)
```

```python
X_traformed_df = pd.DataFrame(X_traformed,columns=trf1.get_feature_names_out())
```

```python
X_traformed_df.head()
```

```python
# OHE --> embarked,sex
trf2 = ColumnTransformer(transformers=[
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,3])
],remainder='passthrough')
```

```python
X_transformed = trf2.fit_transform(X_traformed)
X_transformed_df = pd.DataFrame(X_transformed , columns=trf2.get_feature_names_out())
X_transformed_df
```

```python
# Scalling
trf3 = ColumnTransformer(transformers=[
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
#feture selection
trf4 = SelectKBest(score_func=chi2,k=5)
```

```python
#train the model
trf5 = DecisionTreeClassifier()
```

# Creating Pipelines

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Alternative Make pipeline

```python
pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
```

```python
pipe.fit(X_train,y_train)
```

# Exploring Pipeline

```python
pipe.named_steps
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
pipe.named_steps['trf1'].transformers_[1][1].statistics_
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

# Cross Validation using Pipelines

```python
from sklearn.model_selection import cross_val_score

cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

# GridSearch using Pipelines

```python
params={
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe,params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
