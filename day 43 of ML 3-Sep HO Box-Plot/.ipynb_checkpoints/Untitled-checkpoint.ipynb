```python
#finding IQR

percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile75,percentile25
```

```python
iqr = percentile75-percentile25
iqr
```

```python
# finding boundary
highestBoundary = percentile75+1.5*iqr
lowestBoundary = percentile25 - 1.5*iqr
```

```python
highestBoundary,lowestBoundary
```

# Finding OUtlier

```python
df[(df['placement_exam_marks'] > highestBoundary ) | (df['placement_exam_marks'] < lowestBoundary)]
```

# Trimming

```python
new_df = df[(df['placement_exam_marks'] < highestBoundary ) & (df['placement_exam_marks'] > lowestBoundary)]
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df['placement_exam_marks'])
             
plt.subplot(1,2,2)
sns.distplot(new_df['placement_exam_marks'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(df['placement_exam_marks'])
             
plt.subplot(1,2,2)
sns.boxplot(new_df['placement_exam_marks'])
plt.show()
```

# Capping

```python
new_df = df.copy()
new_df['placement_exam_marks'] = np.where(
    df['placement_exam_marks'] > highestBoundary,
    highestBoundary, 
    np.where(
        df['placement_exam_marks'] < lowestBoundary,
        lowestBoundary,
        df['placement_exam_marks']
    )
)
```

```python
new_df.shape
```

```python
new_df['placement_exam_marks'].describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])
             
plt.subplot(2,2,2)
sns.distplot(new_df['placement_exam_marks'])

plt.figure(figsize=(12,5))
plt.subplot(2,2,3)
sns.boxplot(df['placement_exam_marks'])
             
plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])
plt.show()
```
