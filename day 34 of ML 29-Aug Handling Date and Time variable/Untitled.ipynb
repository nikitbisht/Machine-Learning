```python
# converting date time datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# Extract year

```python
date['d_year'] = date['date'].dt.year
```

```python
date.head()
```

# Extract month

```python
date['d_month'] = date['date'].dt.month
```

```python
date.head()
```

```python
date['d_m_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

# Extract day

```python
date['d_day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
#weekend
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

# EXtract week of the year

```python
date['date_week'] = date['date'].dt.week
date.head()
```

# EXtract Quater

```python
date['Quater'] = date['date'].dt.quarter
date.head()
```
